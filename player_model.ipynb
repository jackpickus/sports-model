```python
import datetime

for i in range(1, len(df_player2)):
    prev_date = str(df_player2.loc[i-1, 'date'])
    curr_date = str(df_player2.loc[i, 'date'])
    d1 = datetime.datetime.strptime(prev_date, "%Y-%m-%d")
    d2 = datetime.datetime.strptime(curr_date, "%Y-%m-%d")
    df_player2.loc[i, 'Rest'] = abs((d2 - d1).days)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(max_leaf_nodes=25, random_state=1, max_depth=10)

rf_model.fit(train_X, train_y)

score_preds = rf_model.predict(val_X)
rf_val_mae = mean_absolute_error(val_y, score_preds)
print(rf_val_mae)
```
